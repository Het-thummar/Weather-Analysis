### pie chart for search hostory

```python
def pie_chart(history_data):
    labels = []
    values = []

    for i,j in history_data.items() :
        labels.append(i)
        values.append(j)

    plt.figure(figsize=(8, 6))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True ,wedgeprops={'edgecolor': 'black'})

    plt.legend()
    plt.show()
```

### method for return weather data for any city

```python
def city_weather_data(city_name) :
    API_KEY = "f9e2a2a95686e8bec9db7465c6fb86eb"
    url = f"https://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        temp = []
        feels_like = []
        temp_min = []
        temp_max = []
        pressure = []
        sea_level = []
        grnd_level = []
        humidity = []
        description = []
        wind_speed = []
        date_time = []
        
        
        for i in data["list"]:
            temp.append(i['main']['temp'])
            feels_like.append(i['main']['feels_like'])
            temp_min.append(i['main']['temp_min'])
            temp_max.append(i['main']['temp_max'])
            pressure.append(i['main']['pressure'])
            sea_level.append(i['main']['sea_level'])
            grnd_level.append(i['main']['grnd_level'])
            humidity.append(i['main']['humidity'])
            description.append(i['weather'][0]['description'])
            wind_speed.append(i['wind']['speed'])
            date_time.append(i['dt_txt'])
    
    
        data = {'temp':temp,'feels_like':feels_like,'temp_min':temp_min,'temp_max':temp_max,'pressure':pressure,'sea_level':sea_level,'grnd_level':grnd_level,'humidity':humidity,'description':description,'wind_speed':wind_speed,'date_time':date_time}
        df = pd.DataFrame(data)
        df['date_time']=pd.to_datetime(df['date_time'])
        df = df.set_index(['date_time'])
    
        df.to_csv('Weather_data.csv')
        weather_data = pd.read_csv("Weather_data.csv")
        return weather_data
    
    else :
        print("Please enter valid city name")
```
